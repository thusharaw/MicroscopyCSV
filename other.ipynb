```python
import os
import pandas as pd

def strip(x):
    return x.strip() if type(x) == str else x

file1="/allen/aics/microscopy/PRODUCTION/PipelineMetaData/PlateLayout.csv"
df_plate_layout = pd.read_csv(file1, sep=',', header=0, names=['PlateLayoutID','column_id','row_id','well_id','contents','concentration','CellLine','clone_id','well_type','dyes_applied'])
df_plate_layout = df_plate_layout.applymap(strip)
df_plate_layout['well_id'] = df_plate_layout['well_id'].map(lambda x: x[0] + str(x[1:]).zfill(2))
df_plate_layout
```

```python
import math
file2="/allen/aics/microscopy/PRODUCTION/PipelineMetaData/Pipelines.csv"
df_pipeline = pd.read_csv(file2, sep=',', header=0, names=['PipelineID','ExperimentID','PlateID','PlateLayoutID','SplitDate(Pipeline Exp label)','DescriptionPipeline','DescriptionExperiment','WebLink','ChosenForPlating'])
df_pipeline = df_pipeline.applymap(strip)
df_pipeline['PlateID'] = df_pipeline['PlateID'].map(lambda x: 0 if math.isnan(x) else int(x))
df_pipeline['SplitDate(Pipeline Exp label)'] = df_pipeline['SplitDate(Pipeline Exp label)'].map(lambda x: str(x).strip())
del(df_pipeline['ChosenForPlating'])
df_pipeline
```

```python
df_pipeline_plate_layout = pd.merge(df_plate_layout, df_pipeline, on='PlateLayoutID')
#print (df_pipeline_plate_layout['PlateID'].map(lambda x: type(x)))
print (df_pipeline_plate_layout['well_id'].map(lambda x: (x,len(x))))
sr = df_pipeline_plate_layout['PlateLayoutID']
sr[sr == 'L3500001098']
```

```python
image_dir = "/allen/aics/microscopy/PRODUCTION/PIPELINE_4_OptimizationAutomation/3500001101/ZSD1/100X_zstack/"
import glob
[os.path.basename(p) for p in glob.glob("%s*.czi" % image_dir) ]
```

```python
import re
def plate_well_from_czi(fname):
    prog = re.compile('(\d+)_(\d+)X_(\d+)_(\d+)-Scene-(\d+)-P(\d+)-([A-Z]\d+)\.czi')
    sr = prog.match(fname)
    return (sr.group(1), sr.group(7))

plate_well_from_czi('3500001101_100X_20170722_1-Scene-17-P18-F04.czi')
```

```python
def three_tuple(fname, plate_well_pair):
    return (fname, plate_well_pair[0], plate_well_pair[1])

file_plate_well_list = [three_tuple(os.path.basename(p), plate_well_from_czi(os.path.basename(p))) for p in glob.glob("%s*.czi" % image_dir) ]
file_plate_well_list
```

```python
import numpy as np
file_names = [f for f,p,w in file_plate_well_list]
plate_ids = [np.int64(p) for f,p,w in file_plate_well_list]
well_ids = [w for f,p,w in file_plate_well_list]
df_images = pd.DataFrame({'FileName': file_names, 'PlateID': plate_ids, 'WellID': well_ids})
df_images
```

```python
import math

def try_int(x, default=0):
    try:
        return int(x)
    except ValueError:
        return default

df_image_pipeline_plate_layout = pd.merge(df_images, df_pipeline_plate_layout, how='left', left_on = ['PlateID','WellID'], right_on=['PlateID','well_id'])
#df_image_pipeline_plate_layout = pd.merge(df_images, df_pipeline_plate_layout, how='left', left_on = ['PlateID'], right_on=['PlateID'])
#df_image_pipeline_plate_layout = pd.merge(df_images, df_pipeline_plate_layout, how='left', left_on = ['WellID'], right_on=['well_id'])

df_image_pipeline_plate_layout['CellLine'] = df_image_pipeline_plate_layout['CellLine'].map(lambda x: 0 if math.isnan(x) else int(x))
df_image_pipeline_plate_layout['clone_id'] = df_image_pipeline_plate_layout['clone_id'].map(lambda x: try_int(x))
df_image_pipeline_plate_layout
```
