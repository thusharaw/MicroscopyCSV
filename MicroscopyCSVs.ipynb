```python
import os
import pandas as pd
file1="/allen/aics/microscopy/PRODUCTION/PipelineMetaData/PlateLayout.csv"
df_plate_layout = pd.read_csv(file1, sep=',', header=0, names=['PlateLayoutID','column_id','row_id','well_id','contents','concentration','CellLine','clone_id','well_type','dyes_applied'])
df_plate_layout
```

```python
file2="/allen/aics/microscopy/PRODUCTION/PipelineMetaData/Pipelines.csv"
df_pipeline = pd.read_csv(file2, sep=',', header=0, names=['PipelineID','ExperimentID','PlateID','PlateLayoutID','SplitDate(Pipeline Exp label)','DescriptionPipeline','DescriptionExperiment','WebLink','ChosenForPlating'])
df_pipeline
```

```python
df_pipeline_plate_layout = pd.merge(df_plate_layout, df_pipeline, on='PlateLayoutID')
df_pipeline_plate_layout
```

```python
image_dir = "/allen/aics/microscopy/PRODUCTION/PIPELINE_Pilot_09292016/3500000971/3500000971_Scan 6-13-2017 6-17-12 AM_Export 6-13-2017 7-22-28 AM/"
import glob
import os
[os.path.basename(p) for p in glob.glob("%s*.tiff" % image_dir) ]
```

```python
import re
prog = re.compile('(\d+)_SA_([A-Z])_(\d+)_Ch1\.tiff')
def plate_well(fname):
    sr = prog.match(fname)
    return (sr.group(1), sr.group(2)+sr.group(3))
plate_well('3500000971_SA_F_8_Ch1.tiff')
```

```python
def three_tuple(fname, plate_well_pair):
    return (fname, plate_well_pair[0], plate_well_pair[1])

file_plate_well_list = [three_tuple(os.path.basename(p), plate_well(os.path.basename(p))) for p in glob.glob("%s*.tiff" % image_dir) ]
file_plate_well_list
```

```python
import numpy as np
file_names = [f for f,p,w in file_plate_well_list]
plate_ids = [np.int64(p) for f,p,w in file_plate_well_list]
well_ids = [w for f,p,w in file_plate_well_list]
df_images = pd.DataFrame({'FileName': file_names, 'PlateID': plate_ids, 'WellID': well_ids})
df_images
```

```python
df_image_pipeline_plate_layout = pd.merge(df_images, df_pipeline_plate_layout, how='left', left_on = ['PlateID','WellID'], right_on=['PlateID','well_id'])
df_image_pipeline_plate_layout['CellLine'] = df_image_pipeline_plate_layout['CellLine'].map(lambda x: int(x))
df_image_pipeline_plate_layout['clone_id'] = df_image_pipeline_plate_layout['clone_id'].map(lambda x: int(x))
df_image_pipeline_plate_layout
```

```python
import math
file3="/allen/aics/microscopy/PRODUCTION/PipelineMetaData/SourcePlates.csv"
df_source_plates = pd.read_csv(file3, sep=',', header=0, names=['ExperimentID','SourcePlateID','Description','CellLine','clone_id','PassageNumber','DonorPassageNumber','PlateType','Maintainer','CoatingType','CoatingLot','CoatingDate','CoatingMethod','CellBatch','SourceVial','SourcePlate','CellCount (Vicell)','SeedingDensity','CellViability','SeedDate','FeedDate','SplitDate(Pipeline Exp label)','GE Score','GE_Initial','M_Score','M_Initial','Notes','ChosenForPlating','ImageLink','y','z'])
df_source_plates['ExperimentID'] = df_source_plates['ExperimentID'].map(lambda x: x.lower())
df_source_plates['CellLine'] = df_source_plates['CellLine'].map(lambda x: 0 if math.isnan(x) else int(x))
df_source_plates['clone_id'] = df_source_plates['clone_id'].map(lambda x: 0 if type(x) == str or math.isnan(x) else int(x))
df_source_plates
```

```python
df7 = pd.merge(df_image_pipeline_plate_layout, df_source_plates, how='left', 
    left_on = ['CellLine', 'clone_id', 'SplitDate(Pipeline Exp label)'], 
    right_on= ['CellLine', 'clone_id', 'SplitDate(Pipeline Exp label)'])
df7 = df7[(df7['GE Score'] == "Pass")]
result = df7[['FileName','PlateID', 'PassageNumber', 'WellID','CellLine', 'clone_id', 'PlateType','concentration', 
     'Maintainer', 'GE_Initial', 'CellCount (Vicell)', 
     'SplitDate(Pipeline Exp label)', 'SeedDate', 'well_type']]
print (result)
#df7
result.to_csv('/home/thusharaw/imagemetadata.csv')
```

```python
r = result.apply(lambda row: { 
                                row['FileName'] : { 
                                    "cell_info" : {
                                        "line": row['CellLine'],
                                        "clone": row['clone_id'],
                                        "passage": row['PassageNumber']
                                    },
                                    "microscopy_info" : {
                                        "omero_project_name": row['CellLine'],
                                        "omero_dataset_name": row['PlateID'],
                                        "well": row['WellID'],
                                        "concentration": row['concentration'],
                                        "maintainer" :row['Maintainer']
                                    } 
                                    
                                }  
                             } , axis=1)
import json
l = json.dumps(list(r.to_dict().values()))
with open('/home/thusharaw/imagemetadata.json', 'w+') as f:
    f.write(l)

#l = result.to_dict(orient='records')
#import json
#json.dumps(l)
```
